```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the csv file and drop unused columns
trees = pd.read_csv('trees.csv')
trees.drop(columns=["OBJECTID"], inplace=True)
trees.drop(columns=["TYPE"], inplace=True)

# Instantiate a 1000 x 1000 ndarray of zeros to fill in
gps_array = np.zeros((1001, 1001), dtype=int)

# Iterate over every row, calculating the matching x and y indices from the longitude and latitude of that row and then adding one to the specified location in the array
for index, row in trees.iterrows():
    x_loc = -(-71.2 - (row['X'])) / 0.0003
    y_loc = -((row['Y']) - 42.2) / 0.0002
    gps_array[int(y_loc)][int(x_loc)] += 1 

# Simple visualization to show success! Set max to 1 to improve readability. 
plt.imshow(gps_array, cmap='Greys', vmax=1)
plt.title("Array of Boston's Trees")
plt.ylabel('Latitude 42.4 to 42.2')
plt.xlabel('Longitude -71.2 to -70.9')
plt.show()
```

```python
# Guiding scatterplot I made early on to help with visualization
plt.figure(dpi=100)
plt.scatter(trees.X, trees.Y, marker='.', s=0.1, color='green', label="Trees")
plt.xlabel('This is the j-axis, longitude')
plt.ylabel('This is the i-ais, latitude')
plt.title("Trees in Boston")
plt.show()
```
