```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
wifi = pd.read_csv('boston_wifi.csv')
cannabis = pd.read_csv('boston_cannabis_registry.csv')
cannabis = cannabis[(cannabis['Application_Status'] == 'Approved') | (cannabis['Application_Status'] == 'Open for Operations')]
```

```python
con = sqlite3.connect('HW2.db')
query = "SELECT Lat, Long FROM wifi WHERE  Lat like '%.%' and Long like '%.%'"
wifi = pd.read_sql_query(query, con)
```

```python
plt.figure(dpi=100)
#plt.scatter(crime_22.Long, crime_22.Lat, marker='.', s=0.01, color="Red", label="2022 Crimes")
plt.scatter(cannabis.x, cannabis.y, marker='.', color="Black", label="Cannabis Dispensaries")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Cannabis Dispensaries in Boston")
plt.legend(loc='lower right')
plt.show()
```
