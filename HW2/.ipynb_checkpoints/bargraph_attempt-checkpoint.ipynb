```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
con = sqlite3.connect('HW2.db')
query = "SELECT OFFENSE_DESCRIPTION, COUNT(*) AS Count_2018 FROM crime_2018 GROUP BY OFFENSE_DESCRIPTION"
df1 = pd.read_sql_query(query, con)

query = "SELECT OFFENSE_DESCRIPTION, COUNT(*) AS Count_2020 FROM crime_2020 GROUP BY OFFENSE_DESCRIPTION"
df2 = pd.read_sql_query(query, con)

query = "SELECT OFFENSE_DESCRIPTION, COUNT(*) AS Count_2022 FROM crime_2022 GROUP BY OFFENSE_DESCRIPTION"
df3 = pd.read_sql_query(query, con)
```

```python
merged_df = pd.merge(df1, df2, on='OFFENSE_DESCRIPTION', how='outer', suffixes=('_year1', '_year2'))
merged_df = pd.merge(merged_df, df3, on='OFFENSE_DESCRIPTION', how='outer')
merged_df.set_index('OFFENSE_DESCRIPTION', inplace=True)
```

```python
df1.plot(kind='bar', stacked=True)
plt.xlabel('Offense Description')
plt.ylabel('Count')
plt.title('Overlapping Bar Graph of Crimes')
plt.legend(['2018'])
plt.show()
```

```python
df2.plot(kind='bar', stacked=True)
plt.xlabel('Offense Description')
plt.ylabel('Count')
plt.title('Overlapping Bar Graph of Crimes')
plt.legend(['2020'])
plt.show()
```

```python
df3.plot(kind='bar', stacked=True)
plt.xlabel('Offense Description')
plt.ylabel('Count')
plt.title('Overlapping Bar Graph of Crimes')
plt.legend(['2022'])
plt.show()
```

```python
merged_df.plot(kind='bar', stacked=True)
plt.xlabel('Offense Description')
plt.ylabel('Count')
plt.title('Overlapping Bar Graph of Crimes')
plt.legend(['2018', '2020', '2022'])
plt.show()
```
