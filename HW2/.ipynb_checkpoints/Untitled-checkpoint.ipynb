```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
con = sqlite3.connect('HW2.db')
```

```python
query_18 = """
SELECT *
FROM crime_2018
WHERE Lat between 42.2 and 42.4 
AND Long between -70.9 and -71.2
"""
db_18 = pd.read_sql_query(query_18, con)
db_18
```

```python
query_20 = """
SELECT *
FROM crime_2020
WHERE Lat between 42.2 and 42.4 
AND Long between -70.9 and -71.2
"""
db_20 = pd.read_sql_query(query_20, con)
db_20
```

```python
query_22 = """
SELECT *
FROM crime_2022
WHERE Lat between 42.2 and 42.4 
AND Long between -70.9 and -71.2
"""
db_22 = pd.read_sql_query(query_22, con)
db_22
```
