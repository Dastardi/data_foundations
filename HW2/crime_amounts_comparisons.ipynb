```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
con = sqlite3.connect('HW2.db')
```

```python
query_18 = """
SELECT OFFENSE_DESCRIPTION, COUNT(*) AS Count_2018
FROM crime_2018
WHERE lat between 42.2 and 42.4 
AND lon between -71.2 and -70.9
GROUP BY OFFENSE_DESCRIPTION
"""
```

```python
query_20 = """
SELECT OFFENSE_DESCRIPTION, COUNT(*) AS Count_2020
FROM crime_2020
WHERE lat between 42.2 and 42.4 
AND lon between -71.2 and -70.9
GROUP BY OFFENSE_DESCRIPTION
"""
```

```python
query_22 = """
SELECT OFFENSE_DESCRIPTION, COUNT(*) AS Count_2022
FROM crime_2022
WHERE lat between 42.2 and 42.4 
AND lon between -71.2 and -70.9
GROUP BY OFFENSE_DESCRIPTION
"""
```

```python
db_18 = pd.read_sql_query(query_18, con)
db_18
```

```python
db_20 = pd.read_sql_query(query_20, con)
```

```python
db_22 = pd.read_sql_query(query_22, con)
```
