```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
con = sqlite3.connect('HW2.db')
```

```python
query = "select Application_Status, x, y from cannabis where Application_Status != 'Withdrawn'"
cannabis = pd.read_sql_query(query, con)
cannabis
```

```python
crime_22 = pd.read_csv('boston_crime_2022.csv', low_memory=False)
crime_22 = crime_22[(crime_22['Lat'] >= 42.2) & (crime_22['Lat'] <= 42.4)]
crime_22 = crime_22[(crime_22['Long'] >= -71.2) & (crime_22['Long'] <= -71)]
crime_22
```

```python
crime_loc = [Point(xy) for xy in zip(crime_22['Long'], crime_22['Lat'])]
crime_gdf = gpd.GeoDataFrame(crime_22, geometry=crime_loc)
crime_gdf
```

```python
cannabis_loc = [Point(xy) for xy in zip(cannabis['x'], cannabis['y'])]
cannabis_gdf = gpd.GeoDataFrame(cannabis, geometry=cannabis_loc)
cannabis_gdf
```
