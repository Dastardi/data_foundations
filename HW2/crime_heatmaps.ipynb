```python
import matplotlib.pyplot as plt
import pandas as pd
wifi = pd.read_csv('boston_wifi.csv')
```

```python
crime_18 = pd.read_csv('boston_crime_2018.csv')
```

```python
crime_20 = pd.read_csv('boston_crime_2020.csv', low_memory=False)
crime_20 = crime_20[(crime_20['Lat'] >= 42.2) & (crime_20['Lat'] <= 42.4)]
crime_20 = crime_20[(crime_20['Long'] >= -71.2) & (crime_20['Long'] <= -71)]
```

```python
crime_22 = pd.read_csv('boston_crime_2022.csv', low_memory=False)
crime_22 = crime_22[(crime_22['Lat'] >= 42.2) & (crime_22['Lat'] <= 42.4)]
crime_22 = crime_22[(crime_22['Long'] >= -71.2) & (crime_22['Long'] <= -71)]
```

```python
plt.figure(dpi=200, figsize=(7.5,5))
plt.hexbin(crime_18.Long, crime_18.Lat, gridsize=35, cmap='cividis', vmax=500, bins=None)
plt.axis([crime_18.Long.min(), crime_18.Long.max(), crime_18.Lat.min(), crime_18.Lat.max()])

cbar = plt.colorbar(label="Number of Crimes", ticks=[0, 100, 200, 300, 400, 500])
cbar.ax.set_yticklabels(['0','100','200','300','400','500+'])

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Crimes in Boston (2018)")
plt.show()  
```

```python
plt.figure(dpi=200, figsize=(7.5,5))
plt.hexbin(crime_20.Long, crime_20.Lat, gridsize=35, cmap='cividis', vmax=500, bins=None)
plt.axis([crime_20.Long.min(), crime_20.Long.max(), crime_20.Lat.min(), crime_20.Lat.max()])

cbar = plt.colorbar(label="Number of Crimes", ticks=[0, 100, 200, 300, 400, 500])
cbar.ax.set_yticklabels(['0','100','200','300','400','500+'])

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Crimes in Boston (2020)")
plt.show()  
```
